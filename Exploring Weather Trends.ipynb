## Step 4: Data Visualization using Python

To analyse the __'San Francisco' data__ and __Global data__ I will be using python. 

Import necessary libraries.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

__San Francisco data__ is stored in variable __sf__.

````python
sf = pd.read_csv('San Francisco.csv')
````

__Global data__ is stored in variable __g__ .

````python
g = pd.read_csv('global.csv')
````

sf.head(20) code shows the first 20 rows of San Francisco city weather data.

````python
sf.head(20)
````

### Global Weather Trends:

The graph below shows the global weather trend over the years for 5-Year Moving Average(blue), 7-Year Moving Average(yellow) and 10-Year Moving Average(green).  

From the graph we can observe the following:
1. The global temperature reached a minimum of average of 7.2ºC between 1810 and 1820.
2. Between 1800 and 1850 the global average temperatue fluctuated drastically.
3. From 1900 the global average temperatue has gradually increased and reached a maximum of 9.7ºC

````python
fig, ax = plt.subplots(figsize=(14,6))
plt.plot(g['year'], g['5-Year MA'],color='blue', label='5-Year MA')
plt.plot(g['year'], g['7-Year MA'],color='orange', label='7-Year MA')
plt.plot(g['year'], g['10-Year MA'],color='green', label='10-Year MA')
plt.title('Global Weather Trends', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temperature(ºC)', fontsize=14)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)
plt.grid()
plt.show()
````

### San Francisco Weather Trends:

The graph below shows the San Francisco weather trend over the years for 5-Year Moving Average(blue), 7-Year Moving Average(yellow), 10-Year Moving Average(green) and 20-Year Moving Average(red).  

From the graph we can observe the following:
1. The moving average temperatue for San Francisco over the year ranges between 13.8ºC to 15.3ºC.
2. Between 1860 and 1940 the average temperatur ranges from 13.8ºC to 14.6ºC, but the average temperatue gradually started increasing from 1960.

````python
fig, ax = plt.subplots(figsize=(14,6))
plt.plot(sf['year'], sf['5-Year MA'],color='blue', label='5-Year MA')
plt.plot(sf['year'], sf['7-Year MA'],color='orange', label='7-Year MA')
plt.plot(sf['year'], sf['10-Year MA'],color='green', label='10-Year MA')
plt.plot(sf['year'], sf['20-Year MA'],color='red', label='20-Year MA', marker= 'o')
plt.title('San Francisco moving average temperature over the years', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temparature (ºC)', fontsize=14)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)
plt.grid()
plt.show()
````

### 5 Years Moving average temparature over the years

Graph below shows the 5 Years Moving average temparature over the years for global data and San Francisco data. 

Note: Since the temperature recodings for San Francisco starts from the year 1849 and the global temperature recordings starts from the 1750, there are (1849 - 1750) 99 years of missing average temperature data for the city of San Francisco. The missing value cannot be replaced by random value as it will affect the analysis. Due to which I have done my analysis for years since the data was avilable i.e. from the year 1849 to 2015.  

From the graph we can observe that the average temperature of San Francisco is higher than the average global temperature.


````python
fig, ax = plt.subplots(figsize=(12,6))
plt.plot(g['year'], g['5-Year MA'],color='red', label='Global', marker ='o')
plt.plot(sf['year'], sf['5-Year MA'],color='blue', label='San Francisco' , marker ='o')
plt.title('5 Years Moving average temparature over the years', fontsize=18)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temparature (ºC)', fontsize=14)
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)
plt.xlim(1853,2013)
plt.grid()
plt.show()
````
